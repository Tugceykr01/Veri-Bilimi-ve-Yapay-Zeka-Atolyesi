```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv')
print(df.head())
```

```python
#2
#kadın-erkek sayısına bakalım

cinsiyet_sayilari = df['gender'].value_counts()
print(cinsiyet_sayilari)
```

```python
#3
#kadın-erkek sayısına histogram grafiği ile bakalım

cinsiyet_sayilari.plot(kind='bar', color=['green', 'blue'])
```

```python
#4
#race/ethnicity sütununda kaç farklı grup olduğuna bakalım

# hangi gruplar olduğuna baktık
farkli_gruplar = df['race/ethnicity'].unique()
print(farkli_gruplar)
```

```python
# gruplardan kaç tane olduğuna baktık
grup_sayilari = df['race/ethnicity'].value_counts()
print(grup_sayilari)
```

```python
# kaç grup olduğuna sayısına baktık
grup_sayisi = df['race/ethnicity'].nunique()
print(f"Toplam {grup_sayisi} farklı grup var.")
```

```python
race = df["race/ethnicity"].value_counts()
race
```

```python
#5
#yukarıda bulduğumuzu görselleştirelim

sns.set(style="whitegrid")
grup_sayilari = df['race/ethnicity'].value_counts().reset_index()
grup_sayilari.columns = ['race/ethnicity', 'Sayı']  
plt.figure(figsize=(6,3))
sns.barplot(x='race/ethnicity', y='Sayı', data=grup_sayilari, palette='Set2')
plt.title('Race/Ethnicity Gruplarının Sayısı')
plt.xlabel('Race/Ethnicity Grubu')
plt.ylabel('Sayı')
plt.xticks(rotation=45)  
plt.show()
```

```python
#9
#gender sütunundaki değerler için ortalama math score, reading score, writing score değerlerini bulalım.

# math score hesaplama
df['math score'] = pd.to_numeric(df['math score'], errors='coerce')
df = df.dropna(subset=['math score', 'gender'])
ortalama_math_score = df.groupby('gender')['math score'].mean()
print(ortalama_math_score)
```

```python
# reading score hesaplama

df['reading score'] = pd.to_numeric(df['reading score'], errors='coerce')
df = df.dropna(subset=['reading score', 'gender'])
ortalama_reading_score = df.groupby('gender')['reading score'].mean()
print(ortalama_reading_score)
```

```python
# writing score

df['writing score'] = pd.to_numeric(df['writing score'], errors='coerce')
df = df.dropna(subset=['writing score', 'gender'])
ortalama_writing_score = df.groupby('gender')['writing score'].mean()
print(ortalama_writing_score)
```
